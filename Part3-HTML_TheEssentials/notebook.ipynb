## **Introduction to HTML**

HTML is a markup language. It is not a full programming language.

Lenguage to mark up documents, in this case, webpages.

It was created to hel describe the structure of academic research papers and share them across the internet.

Its goal is to take text context and mark it up in some structured way.

Elements in HTML are represented by tags. Tags are written in angle brackets.

```html
<p> This is a paragraph </p>
```

The first tag is the opening tag, the second tag is the closing tag.

---

## **HTML Paragraphs**

In HTML, a paragraph is defined by the `<p>` tag.

```html
<p> This is a paragraph </p>
```

This can represent a paragraph, a sentence, a line, or any other portion of text. It is a block level element. 

---

## **HTML Headings**

In HTML, there are 6 levels of headings. They are defined by the `<h1>` through `<h6>` tags.

```html
<h1> This is a heading </h1>
```

The `<h1>` tag is the largest heading, and the `<h6>` tag is the smallest heading.

A single page shouldn't have more than one `<h1>` tag. It is used to represent the main heading of the page.

Smaller headings are used to represent subheadings and should be used in order.

## **Chrome Inspector**

The Chrome Inspector is a tool that allows us to inspect the HTML, CSS, and JavaScript of a webpage. By right clicking on an element and selecting "Inspect", we can see the HTML of that element. Then we can see the CSS that is being applied to that element to understand how it is being styled. 

This way, we can see how other websites are built and learn from them. Making it easier to build our own websites and learn how other people are building theirs.

## **HTML Boilerplate**

By simply adding elements of paragraphs and headings, we are creating an invalid HTML document. Let's view the skeleton of an HTML document.

```html
<!DOCTYPE html>
<html>
    <head>
        <title> My First HTML Document </title>
    </head>
    <body>
        <h1> My First Heading </h1>
        <p> My first paragraph </p>
    </body>
</html>
```

The `<!DOCTYPE html>` tag is the document type declaration. It tells the browser that this is an HTML document.

The `<html>` tag is the root element of an HTML document. It wraps all the content on the entire page and is sometimes referred to as the HTML element.

The `<head>` tag contains meta information about the HTML document. It is not displayed on the page, but is displayed in the browser tab.

The `<title>` tag specifies a title for the HTML document. It is displayed on the browser tab.

The `<body>` tag defines the document's body. It contains all the contents of an HTML document, such as headings, paragraphs, images, hyperlinks, tables, lists, etc.

If we write `!` and press tab, we can create a boilerplate HTML document in VSCode.

## **HTML Lists**


In HTML, there are two types of lists: unordered lists and ordered lists.

### **Unordered Lists**

Unordered lists are defined by the `<ul>` tag. Each item in the list is defined by the `<li>` tag.

```html
<ul>
    <li> Item 1 </li>
    <li> Item 2 </li>
    <li> Item 3 </li>
</ul>

<!-- Output -->
- Item 1
- Item 2
- Item 3
```

### **Ordered Lists**

Ordered lists are defined by the `<ol>` tag. Each item in the list is defined by the `<li>` tag.

```html
<ol>
    <li> Item 1 </li>
    <li> Item 2 </li>
    <li> Item 3 </li>
</ol>

<!-- Output -->
1. Item 1
2. Item 2
3. Item 3
```

Inside the list elements, we can add any other HTML elements, such as paragraphs, headings, images, other lists, etc.

## **Anchor Tags**

Anchor elements are used to link to content outside of the current page. They are defined by the `<a>` tag. The `href` attribute specifies the URL of the page we are linking to.

```html
<a href="https://www.google.com"> Google </a>
```

We can also link to other pages on our own website. We can do this by specifying the path to the page we want to link to.

```html
<a href="/about"> About </a>
```

We can also link to other parts of the same page. We can do this by specifying the `id` of the element we want to link to.

```html
<a href="#contact"> Contact </a>
```

We can also link to an email address. We can do this by specifying the `mailto` attribute.

```html
<a href="mailto:
```

## **Image**

The `<img>` tag is used to embed an image in an HTML document. It is a self-closing tag, meaning it does not need a closing tag.

```html
<img src="https://images.unsplash.com/photo-1622831478474-3b7b4b5b2b0b?ixid=MnwxMjA3fDB8MHxlZGl0b3JpYWwtZmVlZHwzNnx8fGVufDB8fHx8&ixlib=rb-1.2.1&auto=format&fit=crop&w=500&q=60" alt="A beautiful landscape">
```

Technically, it is possible to change the width and height of an image using the `width` and `height` attributes. However, it is not recommended to do so. Instead, we should use CSS to style the image.

If our image is located in the same folder as our HTML document, we can simply specify the name of the image in the `src` attribute. If it is located in a different folder, we can specify the path to the image.

```html
<img src="images/landscape.jpg" alt="A beautiful landscape">
```
